```python
import numpy as numpy
import pandas as pd
```

# Examining the data frame 

```python
perscore_df = pd.read_csv("personality_scores.csv" , delimiter = ";")
perscore_df.head(5)
```

# Checking Missing values

```python
perscore_df.isnull().sum()
perscore_df.head(5)
```

# Checking for duplicates

```python
#Checking if there are duplicates

perscore_df["ID"].is_unique
```

```python
#Checking the number of duplicates
perscore_df["ID"].nunique()
```

There are 1555 duplicates in the ID of the dataframe

# Dropping unnecessary columns

```python
for score in perscore_df :
    if "Unnamed:" in score:
        del perscore_df[score]
```

```python
perscore_df.head(5)
```

# Setting the ID column to be the index of the dataframe

```python
pers_index = perscore_df.set_index("ID")
pers_index.head(5)
```

# Asserting new data frame with original dataframe

```python
assert len(perscore_df) == len(pers_index)
```
